```python
x = np.arange(-11, 11, 0.1)
y = np.sin(x)
# print(x)
# print(y)
plt.xlabel("value range")
plt.ylabel('Range of Sin function')
plt.plot(x, y)
plt.plot([-12, 12],[0, 0], 'k-', lw=1,dashes=[2, 2])
# plt.plot([0, 10], [0, 0.8], 'k-', lw=1,dashes=[2, 2])
# plt.grid()
plt.show()
```

```python
x1 = np.arange(-11, 11, 0.1)
y1 = np.cos(x1)
# print(x)
# print(y)
plt.xlabel("value Range")
plt.ylabel('Range of cos function')
plt.plot(x1, y1)
plt.plot([-12, 12],[0, 0], 'k-', lw=1,dashes=[2, 2])
# plt.plot([0, 10], [0, 0.8], 'k-', lw=1,dashes=[2, 2])
# plt.grid()
plt.show()
```
